```python
csv_file = "red_wine.csv"
red_wine_df = pd.read_csv(csv_file)
```

```python
# Example of creating a histogram for 'quality'
plt.figure(figsize=(8, 6))
plt.hist(red_wine_df['quality'], bins=6, alpha=0.7, color='blue')
plt.title('Distribution of Wine Quality')
plt.xlabel('Quality')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.scatter(red_wine_df['alcohol'], red_wine_df['pH'])
plt.title('Alcohol content vs. pH')
plt.xlabel('Alcohol')
plt.ylabel('pH')
plt.show()
```

```python
plt.boxplot(red_wine_df['sulphates'])
plt.title('Sulphates Distribution')
plt.ylabel('Sulphates')
plt.show()
```
